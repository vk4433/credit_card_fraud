#### splitting the data

```python
x = data.drop('Class', axis = 1) 
y=data['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

#### Model Training

```python
model = {
    "LogisticRegression":LogisticRegression()
}
for name,mod in model.items():
    print(f"------{name}-----")
    mod.fit(X_train,y_train)
    n_pred = mod.predict(X_test)
    print(f"\n Accuaracy: {accuracy_score(y_test, n_pred)}") 
    print(f"\n Precision: {precision_score(y_test, n_pred)}") 
    print(f"\n Recall: {recall_score(y_test, n_pred)}") 
    print(f"\n F1 Score: {f1_score(y_test, n_pred)}")
```

##### it's an fraud detection problem we need mainly focus on precision (false positive)

### undersamling

##### we are performing undersamplig which means we  are equaling the number 0 to 1 . usally we may get loss of the data.


```python
fraud = data[data["Class"]==1]
n_fraud = data[data["Class"]==0] # non-fraudulent transactions
```

```python
fraud.shape
```

```python
n_fraud.shape
```

```python
re_n_fraud =n_fraud.sample(len(fraud)) # we are taking sample from notfraud that equals the size of fraud
```

```python
re_n_fraud.shape
```

```python
un_data = pd.concat([fraud,re_n_fraud],ignore_index=True) # concating both the dataframes
```

#### splitting the dataset

```python
x = un_data.drop('Class', axis = 1) 
y=un_data['Class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = {
    "LogisticRegression":LogisticRegression()  ### model training and evoluting for undersampled data
}
for name,mod in model.items():
    print(f"------{name}-----")
    mod.fit(X_train,y_train)
    un_pred = mod.predict(X_test)
    print(f"\n Accuaracy: {accuracy_score(y_test, un_pred)}") 
    print(f"\n Precision: {precision_score(y_test, un_pred)}")
    print(f"\n Recall: {recall_score(y_test, un_pred)}") 
    print(f"\n F1 Score: {f1_score(y_test, un_pred)}")
```

##### by doing undersampling we can look the there may be in increase precision

### over sampling

```python
x = data.drop('Class', axis = 1) 
y=data['Class']
```

```python
x.shape,y.shape
```

```python
x_reshape,y_reshape = SMOTE().fit_resample(x,y) ### we are using SMOTE to oversample the minority class
```

```python
x_reshape.shape , y_reshape.shape
```

```python
y_reshape.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_reshape,y_reshape,test_size=0.2,random_state=42)
```

```python
model = {
    "LogisticRegression":LogisticRegression() ## training and testing the model by oversampling
}
for name,mod in model.items():
    print(f"------{name}-----")
    mod.fit(X_train,y_train)
    os_pred = mod.predict(X_test)
    print(f"\n Accuaracy: {accuracy_score(y_test, os_pred)}") 
    print(f"\n Precision: {precision_score(y_test, os_pred)}")
    print(f"\n Recall: {recall_score(y_test, os_pred)}") 
    print(f"\n F1 Score: {f1_score(y_test, os_pred)}")
```

##### By doing oversampling we can see increse precision 

#### Checking the oversample model with Xgboost

```python
X_train,X_test,y_train,y_test = train_test_split(x_reshape,y_reshape,test_size=0.2,random_state=42)
```

```python
xgc = XGBClassifier(
    objective='binary:logistic',
    max_depth=3,
    learning_rate=0.1,
    n_estimators=100,
    gini=0.3,
    loss='log_loss'
)
```

```python
xgc
```

```python
xgc.fit(X_train,y_train) #model training
```

```python
xg_pred = xgc.predict(X_test) # predicting the labels of the test set
```

#### confusion matrix

```python
confusion_matrix(y_test,xg_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,xg_pred),annot=True,cmap="coolwarm")
```

#### Evoluating the model

```python
print(f"\nAccuracy: {accuracy_score(y_test, xg_pred)}") 
print(f"\nPrecision: {precision_score(y_test, xg_pred)}")
print(f"\nRecall: {recall_score(y_test, xg_pred)}") 
print(f"\nF1 Score: {f1_score(y_test, xg_pred)}")
```

#### feature importance with shap

```python
import shap
explainer = shap.TreeExplainer(xgc)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")
```

#### Hyperparameters by using Grid search

```python
model = XGBClassifier()
params = {
    "n_estimators":[100,150,200,250],
    'learning_rate': [0.01, 0.1, 0.2],
    "max_depth":[3,5,7,9],
    'min_child_weight': [1, 3, 5],
    }
grid_search = GridSearchCV(estimator=model,param_grid=params,scoring="accuracy",cv=5,n_jobs=-1,verbose=2)
```

```python
grid_search
```

```python
grid_search.fit(X_train, y_train)
```

```python
print(f"Best parameters: {grid_search.best_params_}")
print(f"Best score: {grid_search.best_score_}")
```
